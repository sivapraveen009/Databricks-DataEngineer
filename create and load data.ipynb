```
%python

df = spark.read.format("JSON").option("inferSchema","true").load("dbfs:/FileStore/tables/titanic.json")
cols = ["PassengerId","Survived","Pclass","Name","Sex","Age","SibSp","Parch","Ticket","Fare","Cabin","Embarked"]
df = df.selectExpr(cols)
display(df)

df.createOrReplaceTempView("titanic")
```

```
%python

df.write.format("delta").saveAsTable("titanic")
```
